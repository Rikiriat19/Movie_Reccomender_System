```python
#import libraries 
import pandas as pd
import numpy as np
```

```python
#import datasets
credits = pd.read_csv("/content/tmdb_5000_credits.csv")
movies = pd.read_csv("/content/tmdb_5000_movies.csv")
```

```python
#merge both datasets on the basis of their title
movies_new = movies.merge(credits, on ="title")
```

```python
movies_new.info()
```

```python
#filter out only usable coulmns for our approach 
# moviesid , genres , cast, crew, keywprds, overview, title

movies_new = movies_new[["movie_id","title","genres","overview","keywords","cast","crew"]]
```

```python
movies_new.head(1)
```

```python
movies_new.isnull().sum()
```

```python
#drop null values
movies_new.dropna(inplace=True)
```

```python
#check duplicate values 
movies.duplicated().sum()
```

```python
movies_new.isnull().sum()
```

```python
#for converting the dictionary into list for only the name variable

import ast

def convert(obj):
    L= []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_new['genres']= movies_new['genres'].apply(convert)
```

```python
movies_new['keywords']= movies_new['keywords'].apply(convert)
```

```python
#for getting top 3 artist from the list 
def convert3(obj):
    L= []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies_new['cast'] = movies_new['cast'].apply(convert3)
```

```python
#for fetching the director from the dictionary of the Crew label

def fetch_director(obj):
    L= []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies_new['crew'] = movies_new['crew'].apply(fetch_director)
```

```python
movies_new.head()
```

```python
movies_new['overview'] = movies_new['overview'].apply(lambda x: x.split())
movies_new.head(2)
```

```python
#remove whitespace from the names for making tags simpler

movies_new['genres']= movies_new['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies_new['keywords']= movies_new['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies_new['cast']= movies_new['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies_new['crew']= movies_new['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies_new.head()
```

```python
movies_new['tags'] = movies_new['overview']+movies_new['genres']+movies_new['keywords']+movies_new['cast']+movies_new['crew']
movies_new.head()
```

```python
df = movies_new[['movie_id', 'title', 'tags']]
df.head()
```

```python
# now convert tha tags list into string

df['tags']= df['tags'].apply(lambda x:" ".join(x))
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = "english")
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```
